**1.Use Seaborn and Malplotlib to create a line plot visualizing sales data over time**

Import Libraries: Imports pandas, matplotlib, and seaborn for data manipulation and visualization.
Convert to Datetime: Converts the 'ORDERDATE' column to datetime format.
Extract Year and Month: Extracts the year and month from the 'ORDERDATE' column into new columns.
Group and Sum Sales: Groups the DataFrame by year and calculates the total sales for each year.
Create Line Plot: Plots a line graph of total sales per year, with markers at each data point, and adds labels and a title.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert the ORDERDATE column to datetime format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%d/%m/%Y %H:%M')

# Extract year and month from ORDERDATE
df['YEAR'] = df['ORDERDATE'].dt.year
df['MONTH'] = df['ORDERDATE'].dt.month

# Group by year and calculate total sales
sales_per_year = df.groupby('YEAR')['SALES'].sum().reset_index()

# Create the line plot for total sales per year
plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_per_year, x='YEAR', y='SALES', marker='o')
plt.title('Total Sales Per Year')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

**2.Calculate the percentage of sales contributed by each region.**

 Groups the DataFrame df by the 'COUNTRY' column and calculates the total sales for each country, then calculates the percentage of total sales contributed by each region,
 and sorts the regions by their percentage of total sales. The resulting DataFrame sales_by_region is displayed, showing each country's total sales and their corresponding
 percentage of overall sales.


```python
# Group by country and calculate total sales
sales_by_region = df.groupby('COUNTRY')['SALES'].sum().reset_index()

# Calculate the total sales across all regions
total_sales = sales_by_region['SALES'].sum()

# Calculate the percentage of sales contributed by each region
sales_by_region['Percentage'] = (sales_by_region['SALES'] / total_sales) * 100

# Sort the regions by percentage of sales for better visualization
sales_by_region = sales_by_region.sort_values(by='Percentage', ascending=False)

# Display the sales by region with percentages
print("Sales by Region with Percentages:\n", sales_by_region)

sales_by_region
```

Key Insights:

**6.Calculate the percentage of sales contributed by the top-performing products.**

Converting ORDERDATE to Datetime:
The ORDERDATE column is converted to datetime format to facilitate time-based analysis. 

Extracting the Year:
The year is extracted from the ORDERDATE column and stored in a new column named YEAR.

Grouping by Year and Product Line:
Sales data is grouped by both YEAR and PRODUCTLINE, and the total sales for each group are calculated.

Sorting and Picking Top 3 Products for Each Year:
The data is sorted by YEAR and SALES in descending order to rank products by sales for each year.
The top 3 products for each year are selected using the groupby and head functions.

Calculating Total Sales for Each Year:
Calculate the total sales for each year using groupby and sum.

Calculating Sales Contributed by Top-Performing Products:
Calculate the sales contributed by the top-performing products for each year using groupby and sum.

Merging DataFrames:
Merge the total sales and top products sales DataFrames on YEAR.

Calculating the Percentage:
Calculate the percentage of sales contributed by the top-performing products by dividing the top products sales by the total sales for each year and multiplying by 100.

add Codeadd Markdown

```python
# Convert the ORDERDATE column to datetime format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%d/%m/%Y %H:%M')

# Extract the year from ORDERDATE
df['YEAR'] = df['ORDERDATE'].dt.year

# Group by year and product line, and calculate total sales
sales_by_year_product = df.groupby(['YEAR', 'PRODUCTLINE'])['SALES'].sum().reset_index()

# Sort the data by year and sales, then pick the top 3 products for each year
top_products_yearly = sales_by_year_product.sort_values(['YEAR', 'SALES'], ascending=[True, False]).groupby('YEAR').head(3)

# Calculate the total sales for each year
total_sales_by_year = df.groupby('YEAR')['SALES'].sum().reset_index()

# Calculate the sales contributed by the top-performing products for each year
top_products_sales_by_year = top_products_yearly.groupby('YEAR')['SALES'].sum().reset_index()

# Merge the total sales and top products sales dataframes
merged_sales = pd.merge(total_sales_by_year, top_products_sales_by_year, on='YEAR', suffixes=('_Total', '_TopProducts'))

# Calculate the percentage of sales contributed by the top-performing products
merged_sales['Percentage_TopProducts'] = (merged_sales['SALES_TopProducts'] / merged_sales['SALES_Total']) * 100

# Display the results
print("Percentage of Sales Contributed by Top-Performing Products:\n", merged_sales)
```

```python
# Prepare the data for visualization
top_products_yearly['Percentage_TopProducts'] = top_products_yearly.groupby('YEAR')['SALES'].transform(lambda x: x / x.sum() * 100)

# Visualize the percentage of sales contributed by top-performing products with hue
plt.figure(figsize=(12, 8))
sns.barplot(data=top_products_yearly, x='YEAR', y='Percentage_TopProducts', hue='PRODUCTLINE', palette='viridis')
plt.title('Percentage of Sales Contributed by Top-Performing Products Year on Year')
plt.xlabel('Year')
plt.ylabel('Percentage of Total Sales')
plt.ylim(0, 100)
plt.legend(title='Product Line', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```
